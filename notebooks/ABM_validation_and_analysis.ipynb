```python
# ============================================================
# ABM_Validation_and_Analysis.ipynb
# ============================================================

import sys, os, random
import numpy as np

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..', 'src')))

from model import GroupBuyingModel
import config
from utils import (
    plot_system_dynamics,
    run_stability_test,
    run_parameter_sensitivity,
    save_trust_diffusion_gif,
    plot_dual_trust_diffusion
)

# === 初始化随机种子 ===
random.seed(42)
np.random.seed(42)

# === 1️⃣ 基础验证 ===
print("🧩 Running base simulation...")
model = GroupBuyingModel()
for _ in range(config.STEPS):
    model.step()
plot_system_dynamics(model)

# === 2️⃣ 稳定性分析 ===
print("📈 Running stability test...")
run_stability_test()

# === 3️⃣ 参数敏感性分析 ===
print("🔍 Running parameter sensitivity test...")
run_parameter_sensitivity()

# === 4️⃣ 信任传播可视化 ===
print("🎞️ Generating diffusion animation...")
save_trust_diffusion_gif(model)
plot_dual_trust_diffusion(model)

print("\n✅ All validation and analysis complete.")
```
