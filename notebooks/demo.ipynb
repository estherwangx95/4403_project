```python
import sys, os, random, numpy as np, pandas as pd, matplotlib.pyplot as plt

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..', 'src')))
from model import GroupBuyingModel

random.seed(42)
np.random.seed(42)

model = GroupBuyingModel(n_consumers=20, n_leaders=2)
for i in range(5):
    model.step()

avg_trust = [sum(c.trust for c in model.consumers)/len(model.consumers)]
print(f"Average trust after simulation: {avg_trust[-1]:.2f}")

print("\n✅ Final sales record:", model.sales_record)


plt.plot(model.sales_record, marker='o', linewidth=2)
plt.xlabel("Time Step")
plt.ylabel("Total Sales")
plt.title("Community Group Buying Dynamics (Validated Model)")
plt.grid(True)
plt.show()
```
